```python
qb = pd.DataFrame(data=(draft.query('Pos == "QB"')))
```

```python
num_draft = qb.shape[0]

counts = [1] * num_draft

qb['draft_cnt'] = counts
```

```python
qb1 = pd.DataFrame(qb.groupby('Year').count())
```

```python
qb1.reset_index(inplace =True)
```

```python
qb1
```

```python
qb1['Year'].cov(qb1['draft_cnt'])
```

I wanted to see if the number of quarterbacks being drafted is increasing or decreasing over time. I found a negative corolation which means less quarterbacks are being drafted now than back in 1970.
